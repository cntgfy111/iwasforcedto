```python
# Эта функция принимает на вход датафрейм и возвращает датафрейм с процентом пропусков по столбцам или строкам
# Fргумент axis - по умолчанию 0, то есть по столбцам. 
# Если axis = 1, то по строкам. 
def null_pct(df: pd.DataFrame, name: str, axis: int = 0):
    # axis = 0 - по стобцам, axis = 1 - по строкам
    val_length = len(ancestors_new) if axis == 0 else len(ancestors_new.columns)
    # создаем датафрейм с процентом пропусков
    null_pct_df = pd.DataFrame(df.isnull().sum(axis=axis) / val_length)
    # Сохраняем в эксель
    null_pct_df.to_excel(name+"_missing.xlsx")
    # Возвращаем датафрейм
    return null_pct_df

# Применяем функцию к датафрейму, чтобы посмотреть процент пропусков по столбцам, аргумент axis = 0 по умолчанию
df_missing = null_pct(df, "ancestors_new")
df_missing
```

```python
# Применяем функцию к датафрейму, чтобы посмотреть процент пропусков по строкам, аргумент axis = 1
df_missing_rows = null_pct(df, "ancestors_new", axis=1)
df_missing_rows
```

```python
import tensorflow as tf


SHUFFLE_BUFFER = 500
BATCH_SIZE = 2
print(tf.__version__)
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(X)
```

```python
def as_df(tensor: tf.Tensor):
    return pd.DataFrame(tensor, columns=X.columns)

as_df(normalizer(X.iloc[:3]))
```

```python
from tensorflow import keras
from tensorflow.keras import layers

def r2(y, y_pred):
    unexplained_error = tf.reduce_sum(tf.square(tf.subtract(y, y_pred)))
    total_error = tf.reduce_sum(tf.square(tf.subtract(y, tf.reduce_mean(y))))
    R_squared = tf.subtract(1.0, tf.truediv(unexplained_error, total_error))

    return R_squared

model = keras.Sequential([
    normalizer,
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])
model.compile(loss='mean_absolute_percentage_error', 
              optimizer=keras.optimizers.legacy.Adam(0.001),
              metrics=[r2]
              )
```

```python
model.summary()
```

```python
history = model.fit(
    X,
    y,
    validation_split=0.2,
    epochs=100
)
```

```python
from matplotlib import pyplot as plt

def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.grid(True)

plot_loss(history)
```
